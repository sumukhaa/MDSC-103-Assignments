**Importing the required libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Constructing a while loop and coding gradient descent to make it work for y=x^2**

```python
f = lambda x: x*x
x = np.linspace(-5,5,100)
plt.plot(x,f(x))
plt.title(f'The function $f(x)=x^2$')

arr_x=[]
x,i,n=4,0,0.1
err=1e-5

plt.plot(x,f(x),marker='*',c='r',markersize=10)

while i<100:
    y=x-(n*2*x)
    
    if(np.abs(y-x)<err):
        break
    
    arr_x.append(x)
    x=y
    i+=1

for x in arr_x:
    plt.scatter(x,f(x))
    
#Printing x-axis
plt.axhline()
#Printing y-axis
plt.axvline()

#Bolding the x and y axes
plt.axhline(0,c='k')
plt.axvline(0,c='k')

plt.xlabel('X values in GD Method')
plt.ylabel('Y values in GD Method')

plt.grid()

plt.show()

plt.show()
```

**Defining a gradient descent function** 

```python
#x is initial point, n is learning rate, e is tolerance of the values
def grad_desc(func,df,x,n,e,iterations):
    arr=[]
    i=0 
    while i<iters:
        y=x-(n*df(x))
        arr.append(y)
        if (np.abs(y-x))<e:
            break
        x=y
        i+=1        
                
    return arr
```

**User can provide his own function, its gradient and other parameters here**

```python
f = lambda x: x**3          #f is the function
df = lambda x: 3*x**2       #df is its gradient
x = 4                       #x is initial point
n = 0.01                    #n is learning rate
e = 1e-5                    #e is tolerance of the values
iters = 100
```

**Plotting all the iteration points of GD method**

```python
#Calling the function and storing all x(k+1) values in an array
array=grad_desc(f,df,x,n,e,iters)

#Plotting initial point as a star
plt.plot(x,f(x),marker='*',c='r',markersize=10)

#Iterating thru the array to plot points
for elem in array:
    plt.scatter(elem, f(elem))

pts = np.linspace(-5,5,100)
plt.plot(pts,f(pts),c='y')#olive
plt.title(f'The function $f(x)$')

#Printing x-axis
plt.axhline()
#Printing y-axis
plt.axvline()

#Bolding the x and y axes
plt.axhline(0,c='k')
plt.axvline(0,c='k')

plt.xlabel('X values in GD Method')
plt.ylabel('Y values in GD Method')

plt.grid()

plt.show()
```
