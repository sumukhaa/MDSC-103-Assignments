**Importing the required libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

**Plotting the function:**
$$f(x1,x2) = 4x1 + 6x2 -2x1^2 -2x1x2 -2x2^2$$

```python
#continuous values from 0-50
x=np.linspace(0,50,100)
y=np.linspace(0,50,100)
fig = plt.figure()

#for 3d plotting
ax = fig.add_subplot(projection='3d')

#X and Y are 2D arrays
X, Y = np.meshgrid(x, y)

#Function is defined
Z= 4*X + 6*Y - 2*X*X - 2*X*Y - 2*Y*Y

#Plotting the graph
ax.plot_surface(X,Y,Z, cmap='autumn')

#Labels for the plot
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

**Performing Gradient Ascend on the function**

```python
#No of iterations
iter=100

#Precision for our answer
error=1e-5

#Let origin be start point
start=[0,0]

#Learning rate is 0.01
rate=0.01

#Derivative of the function
derivative= lambda x,y: 4 + 6*Y- 4*X -2*Y -2*Y*Y

#Constructing a while loop to code gradient ascend
i=0
y=[0,0]
while i<iter:
    #we take a point and add the product of rate and partial derivative of the function
    y=y+(rate*derivative(start[0],start[1]))
    
    i+=1
```
